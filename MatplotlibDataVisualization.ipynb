```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
x = [1,2,3,4,5,6]
y = [45,40,20,38,40,50]
plt.plot(x,y)
plt.scatter(x[:30], y[:30], color='r')
plt.xlabel('Day')
plt.ylabel('Temperature')
plt.title('Weather Chart')
```

```python
import numpy as np

x = np.arange(11)

y = x**2

print(x,y)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(x,y, 'g--');
#plt.scatter(x,y);
plt.xlabel('X Axis Title')
plt.ylabel('Y Axis Title')
plt.title('Gragph')
```

```python
plt.subplot(1,3,1)
plt.plot(x,y, 'r*-')

plt.subplot(1,3,2)
plt.plot(y,x, 'b*-')

plt.subplot(1,3,3)
plt.plot(x,y, 'g--')
plt.tight_layout()
```

```python
fig1 =plt.figure(figsize=(12,5))
axes = fig1.add_axes([.1, .1, .9, .5])
axes.plot(x,y, 'b-o')
axes.set_xlabel('X Label')
axes.set_ylabel('y Label');
axes.set_title('Graph')
# axes.set_xlim([1,4])
# axes.set_xlim([1,50])
fig1.savefig('graphsave.png', dpi=200)
```

```python
fig, axes = plt.subplots(nrows = 2, ncols=2)
plt.tight_layout()
```

```python
fig, axes = plt.subplots(nrows = 2, ncols=2, sharex = True, sharey=True)
plt.tight_layout()
axes[0][0].plot(x,y)
axes[0][1].plot(y,x)

```

```python
fig1, axes1 =  plt.subplots(2,1)
fig2, axes2 =  plt.subplots(1,2)
axes1[0].plot(x,y)
axes1[1].plot(y,x)
axes2[0].plot(x,y)
axes2[1].plot(y,x)


```

```python
fig3 =  plt.figure(figsize=(10,5))
axes_main = fig3.add_axes([0.1,0.1,0.9,0.9])
axes_inset = fig3.add_axes([0.3, 0.5, 0.3, 0.4])

axes_main.plot(x,y, 'b')
axes_main.set_xlabel('X label')
axes_main.set_ylabel('X label')
axes_main.set_title('Main_Graph')


axes_inset.plot(y,x, 'r')
axes_inset.set_xlabel('X label')
axes_inset.set_ylabel('X label')
axes_inset.set_title('Inset_Graph')
```
